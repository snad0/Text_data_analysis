Functions

```python
#tokenization function
def tokenize(text):
    text = re.sub(r'[^A-Za-z]',' ',text)
    tokenized_words = word_tokenize(text)
    return tokenized_words

#function to remove stop words
def remove_stopwords(words, stop_words):
    return [x for x in words if x not in stop_words]

#function to count negative and positive score
def negposcount(store, words):
    score = 0
    for x in words:
        if(x in store):
            score = score+1
    return score

#function to calculate polarity according to given formula
def polarity(positive_score, negative_score):
    return (positive_score - negative_score)/((positive_score + negative_score)+ 0.000001)
     
#function to calculate subjectivity according to given formula
def subjectivity(positive_score, negative_score, num_words):
    return (positive_score+negative_score)/(num_words+ 0.000001)

#function to calculate fog_index_cal according to given formula  
def fogindex(average_sentence_length, percentage_complexwords):
    return 0.4*(average_sentence_length + percentage_complexwords)

#function to find complex words
def more_than_2syllables(word):
    if(len(word) > 2 and (word[-2:] == 'es' or word[-2:] == 'ed')):
        return False
    
    w =0
    vowels = ['a','e','i','o','u']
    for i in word:
        if(i.lower() in vowels):
            w = w +1
        
    if(w > 2):
        return True
    else:
        return False
    
    
#syllabul per word
def syllables(word):
    if(word[-2:] == 'es' or word[-2:] == 'ed'):
        return False
    
    w =0
    vowels = ['a','e','i','o','u']
    for i in word:
        if(i.lower() in vowels):
            w = w +1
    if(w > 1):
        return True

def per_pronoun(words):
    pronounRegx=re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
    str_word=" ".join(words)
    pronouns=pronounRegx.findall(str_word)
    num_pronouns= len(pronouns)
    return num_pronouns
```

```python
for i in range(len(reports)):
    content = reports[i]
    if len(content) > 200:
        tokenized_words = tokenize(content)
#         print(len(tokenized_words))
        words = remove_stopwords(tokenized_words, stop_words)
#         print(len(words))
        num_words = len(words)
        positive_score = negposcount(positive_dictionary, words)
#         print(positive_score)
        negative_score = negposcount(Negative_dictionary, words)
        polarity_score = polarity(positive_score, negative_score)
        subjectivity_score = subjectivity(positive_score, negative_score, num_words)
        
        sentences = sent_tokenize(content)
        num_sentences = len(sentences)
        average_sentence_length = num_words/num_sentences
#         print(negative_score)


        num_complexword =0
        uncertainity_score = 0
        constraining_score = 0
        num_syllables=0

        for word in words:
            if(more_than_2syllables(word)):
                num_complexword = num_complexword+1
                
        per_complex_word=(num_complexword/num_words)*100
        
        fog_index = fogindex(average_sentence_length, per_complex_word)
        
        complex_word_count=num_complexword
        
        word_count=num_words
        
        
        for word in words:
            if(syllables(word)):
                num_syllables = num_syllables+1
                
        syllable_per_word=num_syllables/num_words
        
        words_char_len=0
        for word in words:
            char= len(word)
            words_char_len=words_char_len+char
        avg_word_len=words_char_len/word_count
        
        
        
        personal_pronouns=per_pronoun(words)
#         print(positive_score)
        dframe.iloc[[i],[2]]=positive_score
        dframe.iloc[[i],[3]]=negative_score
        dframe.iloc[[i],[4]]=polarity_score
        dframe.iloc[[i],[5]]=subjectivity_score
        dframe.iloc[[i],[6]]=average_sentence_length
        dframe.iloc[[i],[7]]=per_complex_word
        dframe.iloc[[i],[8]]=fog_index
        dframe.iloc[[i],[9]]=average_sentence_length
        dframe.iloc[[i],[10]]=complex_word_count
        dframe.iloc[[i],[11]]=word_count
        dframe.iloc[[i],[12]]=syllable_per_word
        dframe.iloc[[i],[13]]=personal_pronouns
        dframe.iloc[[i],[14]]=avg_word_len


        
```
